# About Columns

### cp : Chest Pain type  
### trtbps : resting blood pressure (in mm Hg)
### fbs : (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)
### chol : cholestoral in mg/dl fetched via BMI sensor
### rest_ecg : resting electrocardiographic results
### thalachh : maximum heart rate achieved
### exng : exercise induced angina (1 = yes; 0 = no)
### oldpeak : Previous peak
### slp : Slope
### caa : number of major vessels (0-4)
### thall : Thal rate
### output : Target variable

```python
plt.figure(figsize=(30,30))
sns.pairplot(df,hue='sex')
```

```python
plt.figure(figsize=(14,12))
sns.heatmap(df.corr(),annot=True)
```

```python
plt.figure(figsize=(12,8))
sns.countplot(x='cp',hue='output',data=df)
plt.title("Chest Pain type",size=20)
plt.xlabel("Chest Pain type",size=10)
plt.ylabel("count",size=10)
plt.grid(axis='y')
```

```python
plt.figure(figsize=(12,8))
sns.countplot(x='fbs',hue='output',data=df)
plt.title("fasting blood sugar",size=20)
plt.xlabel("fasting blood sugar",size=16)
plt.ylabel("count",size=16)
plt.grid()
```

## peoble have less than 120 in fasting blood sugar have more chance of heart attack.

```python
plt.figure(figsize=(12,8))
sns.countplot(x='restecg',hue='output',data=df)
plt.title("Resting electrocardiographic results",size=20)
plt.xlabel("resting electrocardiographic results",size=16)
plt.ylabel("count",size=16)
plt.grid()
```

```python
plt.figure(figsize=(12,8))
sns.countplot(x='caa',hue='output',data=df)
plt.title("number of major vessels",size=20)
plt.xlabel("number of major vessels",size=16)
plt.ylabel("count",size=16)
plt.grid()
```

### peoble having zero of major vessels have more chance of heart attack

```python
plt.figure(figsize=(12,8))
sns.countplot(x='exng',hue='output',data=df)
plt.title("exercise induced angina",size=20)
plt.xlabel("exercise induced angina",size=16)
plt.ylabel("count",size=16)
plt.grid()
```

```python
plt.figure(figsize=(12,8))
sns.countplot(x='sex',hue='output',data=df)
plt.title("Gendre",size=20)
plt.xlabel("Gendre",size=16)
plt.ylabel("count",size=16)
plt.grid()
```

### the chance of heart attack is more in men

```python
plt.figure(figsize=(12,8))
sns.kdeplot(x='age',hue='output',data=df)
plt.title("Age",size=20)
plt.xlabel("age",size=16)
plt.ylabel("count",size=16)
plt.grid()
```

### The chance of heart attack increase from age 30 to age 65.
